**2 LOADING THE LIBRARIES AND DATASET**

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/home/aniekan/Documents/naira performance data.csv')
```

```python
# Viewing the dataset
df
```

The above is the dataset of 3 columns and 15 rows of observations.  The variables are the Date, Naira per U.S dollar, end of period (end of period Rate), and Naira per U.S dollar, period average (Period Average).

The "Naira per U.S. Dollar, end of period" reflects the exchange rate at the close of each month while the "Naira per U.S. Dollar, period average" represents the average exchange rate for the entire month.

**3 VISUALIZING THE DATASET**

```python
# Visualizing how the Naira performs against the Dollar by Naira per U.S. Dollar, end of period against the Date

# Plotting the line graph
plt.plot(df['Date'], df['Naira per U.S. Dollar, end of period'])

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Naira per U.S. Dollar, end of period')
plt.title('Line Graph of Naira per U.S. Dollar, end of period against Date')
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.show()
```

```python
# Visualizing how the Naira performs against the Dollar by Naira per U.S. Dollar, period average against the Date

# Plotting the line graph
plt.plot(df['Date'], df['Naira per U.S. Dollar, period average'])

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Naira per U.S. Dollar, period average')
plt.title('Line Graph of Naira per U.S. Dollar, period average against Date')
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.show()
```

```python
# Visualizing how the Naira performance against the Dollar by Naira per U.S. Dollar, end of period and Dollar by Naira per U.S. Dollar, period average relates

# Plotting the line graph
plt.plot(df['Date'], df['Naira per U.S. Dollar, end of period'], label='Naira per U.S. Dollar, end of period')
plt.plot(df['Date'], df['Naira per U.S. Dollar, period average'], label='Naira per U.S. Dollar, period average')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Naira per U.S. Dollar, period average')
plt.title('Line Graph of Naira per U.S. Dollar, end of period and Naira per U.S. Dollar, period average against Date')
plt.xticks(rotation=45, ha='right')

plt.legend()

# Displaying the plot
plt.show()
```
